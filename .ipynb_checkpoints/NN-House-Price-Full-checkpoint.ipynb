### Random Search

**Customize `neg_rmsle` for Random Search**

‚ö†Ô∏è `GridSearchCV` always tries to _maximize_ a score

üëâ turn `rmsle` into a negative metric to be put into Grid Search

```python
neg_rmsle = make_scorer(lambda y_true, y_pred: mean_squared_log_error(y_true, y_pred, squared=False),
    greater_is_better=False)
```

```python
X_num.shape
```

```python
from scipy import stats
from sklearn.model_selection import RandomizedSearchCV

# Parameter Grid
grid = {
    'max_depth': stats.randint(5,50),
    'min_samples_leaf': stats.randint(10,50),
    'max_features': stats.randint(10,36) # X_num.shape[1] = 36
}

# Random Search
search_tree = RandomizedSearchCV(tree,
                                grid,
                                scoring=neg_rmsle,
                                n_iter=50,
                                cv=5,
                                n_jobs=-1, # Run in parallel
                                verbose=1) # 0: NOT print any log, 1: Summary, 3: Full scores

# Fit data to Random Search
search_tree.fit(X_num, y)
```

```python
search_tree.best_score_ # neg_rmsle for Grid Search
```

```python
search_tree.best_params_
```

```python
print(search_tree.best_estimator_)
```

```python
# Best params
print(f"Best RMSLE: {(-search_tree.best_score_)}\n" # Inverse metric back to positive
      f"Best estimator : {search_tree.best_estimator_}")
```

## Stacking

```python
from sklearn.ensemble import VotingRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor

# Optimal Random Forest
rf = RandomForestRegressor(max_depth=14, 
                                      max_features=25, 
                                      min_samples_leaf=11)

# Optimal AdaBoost
adaboost = AdaBoostRegressor(n_estimators=115,
                             base_estimator=DecisionTreeRegressor(max_depth=28,
                                                                  max_features=13,
                                                                  min_samples_leaf=2))

# Optimal GradientBoosting
gboost = GradientBoostingRegressor(n_estimators=100)

# XGBoost
xgb = XGBRegressor(max_depth=5, 
                   n_estimators=300, 
                   learning_rate=0.05)

# STACKING ALL MODELS TOGETHER
model = VotingRegressor(
    estimators = [("gboost", gboost),
                  ("adaboost", adaboost),
                  ("rf", rf),
                  ("xgboost",xgb)],
    n_jobs=-1
)
```

```python
score = cross_val_score(model,
                       X_num, y,
                       n_jobs=-1,
                       cv=5, scoring=rmsle).mean()

print('RMSLE Score: ', score)
```

## Multimodel Random Search

```python
# Initialze the estimators
tree = DecisionTreeRegressor()
forest = RandomForestRegressor()

# Initiaze the hyperparameters for each dictionary
param1 = {
    'regressor__max_depth': stats.randint(5,50),
    'regressor__min_samples_leaf': stats.randint(10,50),
    'regressor__max_features': stats.randint(10,37),
    'regressor': [tree]
}
param2 = {
    'regressor__max_depth': stats.randint(2,50),
    'regressor__min_samples_leaf': stats.randint(2,50),
    'regressor__max_features': stats.randint(2,37),
    'regressor': [forest]
}

# Make list of dictionaries
params = [param1, param2]
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('regressor', tree)])

# Random Search
multi_search = RandomizedSearchCV(pipeline, #Pipeline
                        params,
                        cv=10,
                        n_jobs=-1,
                        scoring=neg_rmsle)

# Fit data to Random Search
multi_search.fit(X_num,y)
```

```python
# Best params
print(f"Best RMSLE: {(-multi_search.best_score_)}\n" # Inverse metric back to positive
      f"Best estimator : {multi_search.best_estimator_}")
```

## Preprocessing

### Encoding

#### Categorical Feature Encoding

**Import `data-description.csv` as a dictionary**

```python
data_description = pd.read_csv("data-description.csv", delimiter=';')
cat_data_description = data_description.sort_values(by=['Feature']).dropna(axis=0).reset_index(drop=True)
cat_data_description.head(3)
```

```python
cat_names = cat_data_description['Feature'].unique()
cat_names
```

```python
feature_categorical_dict = {}

for i in range(len(cat_names)):
    categories = {}
    feature = cat_names[i]
    filtered_df = cat_data_description[cat_data_description.Feature == feature]
    feature_description = filtered_df.iloc[0]['Feature-Description']
    for index, row in filtered_df.iterrows():
        categories[row['Category']] = row['Category-Description']
    # print(categories)
    feature_categorical_dict[feature] = {
        "0_feature_description": feature_description,
        "1_categories": categories
    }
```

**Create dictionary of all ordinal features**

```python
# In the pipeline for ordinal preprocessing, we'll fill_value='missing' to replace NaN

feat_ordinal_dict = {
    # considers "missing" as "neutral"
    "BsmtCond": ['missing', 'Po', 'Fa', 'TA', 'Gd'], #No Ex
    "BsmtExposure": ['missing', 'No', 'Mn', 'Av', 'Gd'], # 
    "BsmtFinType1": ['missing', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'], #No Partial
    # BsmtFinType1: "BLQ": "Below Average Living Quarters" vs "Rec": "Average Rec Room",
    "BsmtFinType2": ['missing', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'], #
    "BsmtQual": ['missing', 'Fa', 'TA', 'Gd', 'Ex'], #No Po
    "Electrical": ['missing', 'Mix', 'FuseP', 'FuseF', 'FuseA', 'SBrkr'], #
    "ExterCond": ['missing', 'Po', 'Fa', 'TA', 'Gd', 'Ex'], #
    "ExterQual": ['missing', 'Fa', 'TA', 'Gd', 'Ex'], #
    "Fence": ['missing', 'MnWw', 'GdWo', 'MnPrv', 'GdPrv'], #Min Wood/Wire vs Min Privacy 
    "FireplaceQu": ['missing', 'Po', 'Fa', 'TA', 'Gd', 'Ex'], # 
    "Functional": ['missing', 'Sev', 'Maj2', 'Maj1', 'Mod', 'Min2', 'Min1', 'Typ'], #No Sal
    "GarageCond": ['missing', 'Po', 'Fa', 'TA', 'Gd', 'Ex'], #
    "GarageFinish": ['missing', 'Unf', 'RFn', 'Fin'], # 
    "GarageQual": ['missing', 'Po', 'Fa', 'TA', 'Gd', 'Ex'], # 
    "HeatingQC": ['missing', 'Po', 'Fa', 'TA', 'Gd', 'Ex'], #
    "KitchenQual": ['missing', 'Fa', 'TA', 'Gd', 'Ex'], #No Po
    "LandContour": ['missing', 'Low', 'Bnk', 'HLS', 'Lvl'], # 
    "LandSlope": ['missing', 'Sev', 'Mod', 'Gtl'], #
    "LotShape": ['missing', 'IR3', 'IR2', 'IR1', 'Reg'], #
    "PavedDrive": ['missing', 'N', 'P', 'Y'],
    "PoolQC": ['missing', 'Fa', 'Gd', 'Ex'], #No Po
}
```

**Create parameters for [`OrdinalEncoder()`](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html)**

```python
# OrdinalEncoder input (list of string or array-like of intergers or strings)
feat_ordinal = sorted(feat_ordinal_dict.keys()) # sort alphabetically
print(feat_ordinal)
```

```python
# OrdinalEncoder 'categories' parameter (a list of lists/array-like)
feat_ordinal_values_sorted = [feat_ordinal_dict[i] for i in feat_ordinal]
feat_ordinal_values_sorted
```

**Customize `OrdinalEncoder` for pipeline**

```python
#Note: input will be given within the pipeline later
encoder_ordinal = OrdinalEncoder(categories = feat_ordinal_values_sorted,
                                 dtype = np.int64,
                                 handle_unknown = "use_encoded_value", # Use unknown_value below
                                 unknown_value = -1 # Consider unknown values as worse than "missing"
                                )
```

**Full pipeline to preprocess ordinal categorical features**

For every ordinal feature:
- Step 1: Use [`SimpleImpute()`](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html) to replace missing values with 'missing'
- Step 2: Use defined `encoder_ordinal` to encode all ordinal features
- Step 3: `MinMaxScale()` all the encoded values

```python
# In SimpleImputer() set strategy='constant' to replace missing values with fill_value parameter

preproc_ordinal = make_pipeline(SimpleImputer(strategy='constant', fill_value='missing'),
                                encoder_ordinal,
                                MinMaxScaler()
                               )
preproc_ordinal
```

**Ref**: [Missing value estimation methods for DNA microarrays](https://academic.oup.com/bioinformatics/article/17/6/520/272365), 2001.

**Tutorial**: [How to Handle Missing Data with Python and KNN](https://betterdatascience.com/impute-missing-data-with-python-and-knn/)

#### VIF: Measure Multicolinearity

**Check for high multicollinearity features using statsmodels's variance_inflation_factor [`VIF`](https://www.statsmodels.org/dev/generated/statsmodels.stats.outliers_influence.variance_inflation_factor.html)**

VIF: Variance Inflation Factor
- A measure of the amount of multicollinearity per feature
- The higher, the more multicolinear, the less useful the feature with a high VIF...
- Computed by regressing one feature as function of all others features and measuring R-squared

**How to read [`vif_index`](https://www.sigmamagic.com/blogs/what-is-variance-inflation-factor/#:~:text=If%20there%20is%20perfect%20correlation,to%20the%20presence%20of%20multicollinearity.)**
- `inf`: Perfect correlation between two independent variables
- `1`: No multicollinearity
- `4 - 5`: Moderate
- `10 or greater`: Severe

```python
# from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

vif_df = pd.DataFrame()

vif_df["features"] = feat_numerical_scaled.columns

# Loop through feature columns
vif_df["vif_index"] = [vif(feat_numerical_scaled.values, i) for i in range(feat_numerical_scaled.shape[1])]

vif_df.sort_values(by="vif_index", ascending = False)
```

`1stFlrSF, BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, 2ndFlrSF, LowQualFinSF, GrLivArea`

## Create RMSE Score
**Metric**

Submissions are evaluated on Root-Mean-Squared-Error (RMSE) between the logarithm of the predicted value and the logarithm of the observed sales price. (Taking logs means that errors in predicting expensive houses and cheap houses will affect the result equally.)

**Ref**: https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/overview/evaluation


```python
from sklearn.metrics import mean_squared_error
rmse = make_scorer(lambda y_true, y_pred: mean_squared_error(y_true, y_pred)**0.5)

# Create your new scorer to maximize
rmse_neg = make_scorer(lambda y_true, y_pred: -1 * mean_squared_error(y_true, y_pred)**0.5)
```

## Stacking

```python
from sklearn.ensemble import GradientBoostingRegressor

gboost = GradientBoostingRegressor(n_estimators=100)
ridge = Ridge()
svm = SVR(C=1, epsilon=0.05)
adaboost = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=None))


model = VotingRegressor(
    estimators = [("gboost", gboost),("adaboost", adaboost),("ridge", ridge), ("svm_rbf", svm)],
    weights = [1,1,1,1], # to equally weight the two models
    n_jobs=-1
)

pipe_ensemble = make_pipeline(preproc, model, memory=cachedir)

score = cross_val_score(pipe_ensemble, X, y_log, cv=5, scoring=rmse, n_jobs=-1)
score.mean()
```

```python
from sklearn.ensemble import StackingRegressor

gboost = GradientBoostingRegressor(n_estimators=100)
ridge = Ridge()
svm = SVR(C=1, epsilon=0.05)
adaboost = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=None))

model = StackingRegressor(
    estimators=[("gboost", gboost),
                ("adaboost", adaboost),
                ("ridge", ridge),
                ("svm_rbf", svm)],
    final_estimator=LinearRegression(),
    cv=5,
    n_jobs=-1
)

pipe_stacking = make_pipeline(preproc, model, memory=cachedir)
score = cross_val_score(pipe_stacking, X, y_log, cv=5, scoring=rmse, n_jobs=-1)
score.mean()
```
